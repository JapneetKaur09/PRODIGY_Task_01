```python
#department count
department_count = df['Department'].value_counts().reset_index()
department_count
```

```python
department_count.columns = ['Department', 'Count']

# Plotting with seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Count', data=department_count, palette='viridis')
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Number of Employees per Department')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#sns.boxplot(x = 'Age',y = 'MonthlyIncome', data = df, hue = 'Gender')
```

```python
age_gender = df.groupby(['Age', 'Gender']).size().reset_index(name = 'Count')
```

```python
sns.lineplot(x = 'Age', y = 'Count', data = age_gender , hue = 'Gender', marker = 'o')
plt.title('Age and Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
marital_sttsCount = df['MaritalStatus'].value_counts().reset_index()
```

```python
marital_sttsCount.columns = ['MaritalStatus', 'Count']

# Plotting the pie chart
plt.figure(figsize=(5, 5))
plt.pie(marital_sttsCount['Count'], labels=marital_sttsCount['MaritalStatus'], autopct='%1.1f%%', startangle=35, colors=plt.cm.Paired.colors)
plt.title('Distribution of Marital Status')
plt.show()
```

```python
avrg_salary = df.groupby(['Department', 'Gender'])['Salary'].mean().reset_index()
```

```python
age = df[['Age','Gender']]
age
```

```python
plt.figure(figsize=(12, 8))
sns.displot(data=age, x='Age', kind='hist', hue='Gender', palette='viridis', bins=10, aspect=1.5)
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(range(20, 65, 5))    
plt.tight_layout() #adjusts the layout to ensure nothing overlaps and the plot fits well within the figure
plt.show()
```
